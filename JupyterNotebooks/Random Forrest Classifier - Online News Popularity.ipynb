```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/Josiel1998/machine-learning/main/Datasets/OnlineNews-Popularity/OnlineNewsPopularity.csv')
df_one_hot = pd.read_csv('https://raw.githubusercontent.com/Josiel1998/machine-learning/main/Datasets/OnlineNews-Popularity/OnlineNewsPopularity.csv')

df.rename(columns=lambda x: x.strip().lower().replace(" ", ""), inplace=True)
```

```python
df.head()
```

```python
# plot out the Distribution of the dependant varible, shares.
plt.figure(figsize=(20,5))
bin_edges = np.arange(0, df['shares'].max()+1, 200) 
plt.hist(data = df, x = 'shares', bins = bin_edges)
plt.xticks(rotation=90)
plt.title("Distribution of shares")
plt.ylabel("count")
plt.xlabel("shares")
plt.xlim([0, 15000])

plt.vlines(500, 0, 5000, colors = "r",linestyles = "solid")
plt.vlines(3000, 0, 5000, colors = "r",linestyles = "solid")


plt.show()
```
